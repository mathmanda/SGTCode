```python
list5=n_Atlas(5)
list5 = [G for G in list5 if G.num_edges() == 7 and G.clique_number() == 3 and G.is_chordal()]
len(list5)
```

```python
for G in list5:
    L = G.laplacian_matrix()
    evals = L.eigenvalues()
    evals.sort()
    bounds = [2/evals[4], 2/evals[1] ]
    M,r = effresistance_matrix(G, zeroadjacents = False)
    M = 1/r*M
    #print M
    mins = []
    for v in M:
        v = set(v)
        #print v
        v.remove(0)
        #print v
        if len(v) !=0:
            mins+=[min(v)]
    #mins = [min(v) for v in M and  !=0]
    min0 = min(mins)
    maxs = [max(v) for v in M]
    max0 = max(maxs)
    II = allonesmatrix(5) - G.adjacency_matrix()
    Mnon = elementwiseprod(M,II)
    Medge = elementwiseprod(M, G.adjacency_matrix())
    mins = []
    maxs=[]
    for v in Mnon:
        v = set(v)
        v.remove(0)
        #print v
        if len(v) !=0:
            mins+=[min(v)]
            maxs+=[max(v)]
    maxs=[]
    for v in Medge:
        v = set(v)
        v.remove(0)
        #print v
        if len(v) !=0:
            maxs+=[max(v)]
    
    #mins = [min(v) for v in Mnon and v!=0]
    min1 = min(mins)
    #maxs = [max(v) for v in Mnon]
    max1 = max(maxs)
    print '==========='
    G.show()
    print 'diam=', G.diameter(),'lowbound=', bounds[0],'actmin=',min0, 'edgemax=',max1
    print 'nonedgemin=',min1, 'nonedgemax=',max0, 'upbound=', bounds[1]
```
