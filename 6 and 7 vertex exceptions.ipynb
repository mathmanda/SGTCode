```python
bad7 = comparegraphsofsize(7)
```

```python
#How many 7 vertex bad cutoffs have pendants or large simple cycles? 
for G in bad7[0]:
    G.show()
    deg = G.degree()
    if min(deg) == 1:
        print "This graph has a pendant"
    cycles = G.cycle_basis()
    cyclelengths = [len(cycle) for cycle in cycles]
    if max(cyclelengths) >=5:
        print "This graph has a simple cycle of length greater than 4"
    
    print '===Most common Neighbors==='
    print maxcn(G)
    print '===Resistance distances==='
    print lowresist(G,numeric=False)
    
#Answer: Only the first one has a pendant vertex. The first two have simple cycles of length 5.
```

```python
#How many 7 vertex exceptions have pendants or large simple cycles? 
haspend_or_largecycle=[]
pend_largecycle_free=[]
for G in bad7[1]:
    #G.show()
    deg = G.degree()
    cycles = G.cycle_basis()
    cyclelengths = [len(cycle) for cycle in cycles]
    if min(deg) == 1:
        haspend_or_largecycle.append(G)
    elif max(cyclelengths) >=5:
        haspend_or_largecycle.append(G)
    else:
        pend_largecycle_free.append(G)
        
print "7 vertex exceptions with pendant or large cycle:"
print len(haspend_or_largecycle)
print '--------------'
print "7 vertex exceptions without pendant or large cycle:"
print len(pend_largecycle_free)
    
    
#Answer: Only the first one has a pendant vertex. The first two have simple cycles of length 5.
```
